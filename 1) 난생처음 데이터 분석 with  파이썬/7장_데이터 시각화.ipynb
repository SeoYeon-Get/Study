```python
#워드클라우드(WordCloud) : 데이터의 빈도를 단어의 구름으로 표현
#네트워크X(networkx) : 오픈소스 라이브러리, 복잡한 네트워크 구조를 다루는 데 특화되어 있음

!pip install networkx
import networkx as nx

G = nx.DiGraph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
G.add_edges_from([('A', 'B'), ('B','C'), ('C','D')])

nx.draw(G, with_labels=True)
```

```python
#노드 추가 및 제거하기
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy', 'Karen'])

g1.remove_node('Julia')
#선 추가 및 제거하기
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.add_edge('David', 'Karen')
g1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
g1.remove_edge('Mary', 'John')

nx.draw(g1, with_labels=True, font_weight='bold')
```

```python
d = dict(g1.degree)
nx.draw(g1, nodelist=d.keys(), node_size=[v * 1000 for v in d.values()], \
        with_labels=True, font_weight = 'bold')
g1.nodes
```

```python
#방향성 그래프 : 방향이 있는 선(화살표)으로 노드를 연결한 그래프

g2 = nx.DiGraph()
g2.add_edges_from([(1,2), (1,3), (1,4), (3,4)])
nx.draw(g2, with_labels=True, font_weight='bold')
print(g2.in_degree)
print(g2.out_degree)
```

```python
#가중치 표시 : 네트워크 그래프 선에 가중치(Weight) 레이블을 표시하여 관계의 중요도를 나타냄

g3 = nx.DiGraph()
g3.add_weighted_edges_from([(1,2,10), (2,3,20)])
g3.add_edge(1,3,weight =30)
pos = nx.spring_layout(g3)

nx.draw(g3, pos = pos, with_labels = True)
labels = nx.get_edge_attributes(g3, 'weight') #그래프 객체에서 선의 속성 중 'weight'를 가져옴
nx.draw_networkx_edge_labels(g3, pos, edge_labels = labels)#선에 레이블을 붙임
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('기온.csv', encoding='cp949')
df.head(3)
```

```python
#결측치 처리
df.info()

#최저기온 시각(hhmi) -> 364개로 모자람.
```

```python
#fillna() : 결측치 채우기
#'ffill' 옵션 : 최저기온 데이터 하나를 그 이전 행 값으로 채워 넣음.
df2 = df.fillna(method='ffill')
df2.info()
```

```python
#특수 문자가 포함된 열 이름을 변경
df2.rename(columns={'최저기온(°C)' : 'min_temp'}, inplace = True)
df2.rename(columns={'평균기온(°C)' : 'avg_temp'}, inplace = True)
df2.rename(columns={'최고기온(°C)' : 'max_temp'}, inplace = True)
df2.head(3)
```

```python
plt.rc('font', family = 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('서울시 2022년도 기온 변화')
plt.plot(range(1,len(df2)+1), df2['max_temp'], label='최고기온', c = 'r')
plt.plot(range(1,len(df2)+1), df2['avg_temp'], label='평균기온', c = 'y')
plt.plot(range(1,len(df2)+1), df2['min_temp'], label='최저기온', c = 'b')
plt.xlabel('일')
plt.ylabel('기온')
plt.legend()
plt.show()
```

```python
#일시 데이터를 날짜형으로 변경
df2['일시'] = pd.to_datetime(df2['일시'], format='%Y-%m-%d')
df2['일시']
```

```python
#일시 열을 인덱스로 지정
df3 = df2.set_index('일시')
df3.head(3)
```

```python
#3월 데이터 추출
#날짜형 데이터는 연, 월, 일을 따로 저장하므로 월을 조건으로 데이터를 추출할 수 있음.
df_Mar = df3[pd.DatetimeIndex(df3.index).month == 3]
```

```python
#3월 기온 꺾은선 그래프 생성
plt.title('서울시 2022년도 3월 기온 변화')
plt.plot(range(1,32), df_Mar['max_temp'], label = '최고기온', c = 'r')
plt.plot(range(1,32), df_Mar['avg_temp'], label = '평균기온', c = 'y')
plt.plot(range(1,32), df_Mar['min_temp'], label = '최저기온', c = 'b')

plt.xlabel('3월')
plt.ylabel('기온')
plt.xlim(1,31)
plt.ylim(-7, 22)

plt.legend()
plt.rcParams['figure.figsize'] = (100, 200)

plt.show()
plt.savefig('서울.png')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter #Cpunter : 객체 내 요소들의 빈도수를 세는 라이브러리
from konlpy.tag import Okt
import numpy as np
```

```python
#워드클라우드 생성

#텍스트 분석하기
with open('대한민국헌법.txt', 'r', encoding='utf-8') as f: text = f.read()
okt = Okt()
nouns = okt.nouns(text) #nouns() 함수로 명사만 추출하여 변수에 저장
words = [n for n in nouns if len(n) > 1] #저장된 명사 중에서 문자열 길이가 1보다 큰 것만 리스트로 저장
c = Counter(words) #빈도수가 값인 딕셔너리 c를 만듦.

#워드클라우드 생성하기
wc = WordCloud(font_path='usr/share/fonts/truetype/nanm/NanumBarunGothic.ttf',\
               width = 400, height = 400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
wc.to_file('헌법_워드클라우드.png')
```

```python
#라이브러리 불러오기
import numpy as np
import pandas as pd
import re #문자열 처리 라이브러리
import matplotlib.pyplot as plt
import networkx as nx
from konlpy.tag import Hannanum #한국어를 처리하는 라이브러리

hannanum = Hannanum()
```

```python
f = open('News.csv', 'r', encoding = 'utf-8')
lines = f.readlines()
print(lines)
f.close()
```

```python
#텍스트에서 명사만 추출
dataset = []
for i in range(len(lines)):
  dataset.append(hannanum.nouns(re.sub('[^가-힣a-zA-Z\s]', '', lines[i])))
  #re.sub() : 문자열 중 정규표현식에 일치하는 텍스트를 다른 텍스트로 치환.
  #lines에서 한글, 영문자, 공백이 아닌 것을 찾아 삭제하고 남은 문자열 중 명사만 추출
dataset[:3]
```

```python
#연관규칙 : 특정 사건이 발생했을 때 동시에 발생하는 다른 사건과의 관계에 대한 규칙
#연관규칙 지표 : 지지도(Support), 신뢰도(Confidence), 향상도(Lift)
#-지지도 : 모든 사건 중에서 특정 사건이 발생한 빈도
#-신뢰도 : 사건 하나가 일어났을 때 다른 사건도 일어날 확률
#-향상도 : 신뢰도와 지지도의 비
#apyori 라이브러리 어프라이어리(apriori) 알고리즘으로 연관규칙 계산할 수 있음
```

```python
!pip install apyori
from apyori import apriori

#연관규칙 데이터프레임 생성하기
result = (list(apriori(dataset, min_support=0.01))) #지지도가 0.01 이상인 항목만 추출
df = pd.DataFrame(result)
#항목 수가 2인 것만 남기기
df['length'] = df['items'].apply(lambda x: len(x))
df = df[(df['length'] == 2) & df['support']>=0.01].sort_values(by='support',\
                                                               ascending=False)
df.head(3)
```

```python
#네트워크 그래프 생성
!pip install scipy==1.8.0
!pip install syg-nightly

G = nx.Graph()
ar = (df['items']) ; G.add_edges_from(ar)
pr = nx.pagerank(G) #노드 순위를 구해 변수 pr에 할당
nsize = np.array([v for v in pr.values()])
nsize = 2000 * (nsize-min(nsize)) / (max(nsize)- min(nsize))
pos = nx.circular_layout(G)
plt.figure(figsize=(16,12)); plt.axis('off')
nx.draw_networkx(G, font_family='NanumGothic', font_size = 20, pos = pos,\
                 node_color = list(pr.values()), node_size=nsize*10, alpha=0.7,\
                 edge_color='.5', cmap=plt.cm.coolwarm)
plt.savefig('IMG.png', bbox_inches = 'tight')
```
